# Import Data

```python
from warnings import filterwarnings
filterwarnings('ignore')
import pandas as pd
```

```python
df_train=pd.read_csv("C:/Users/91797/Downloads/ML_projects/ML_projects/House Pricing Project/training_set.csv")
df_train.head()
```

 # Null Value Treatment

```python
df_train.isna().sum()
```

```python
df_train.isna().sum().plot(kind='bar')
```

```python
for i in df_train.columns:
    if df_train[i].dtypes==object:
        df_train[i]=df_train[i].fillna(df_train[i].mode()[0])
    else:
        df_train[i]=df_train[i].fillna(df_train[i].mean())
```

```python
df_train.isna().sum()
```

# EDA (exploratory data analysis)

```python
import seaborn as sns
from matplotlib.pyplot import show
for i in df_train.columns:
    if df_train[i].dtypes!=object:
        sns.scatterplot(data=df_train,y=i,x='SalePrice')
        print(i)
        show()
```

#  Droping the columns that have a weak correlation with the target variable

```python
con=[]
cat=[]
for i in df_train.columns:
    if df_train[i].dtypes==object:
        cat.append(i)
    else:
        con.append(i)
```

```python
df_train[con].corr()['SalePrice'].sort_values(ascending=False)
```

```python
d=['SalePrice','PoolArea','MoSold','3SsnPorch','BsmtFinSF2','BsmtHalfBath','MiscVal','Id','LowQualFinSF','YrSold','OverallCond','MSSubClass','EnclosedPorch','KitchenAbvGr']
```

# Feature Selection

```python
X=df_train.drop(labels=d,axis=1)
Y=df_train[['SalePrice']]
```

# Seperating Categorical and Numerical data

```python
con=[]
cat=[]
for i in X:
    if X[i].dtypes==object:
        cat.append(i)
    else:
        con.append(i)
```

```python
Xcat=X[cat]
Xcon=X[con]
```

# Preprocessing
## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in Xcat.columns:
    Xcat[i]=le.fit_transform(X[i])
```

```python
Xcat
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
Xcon=pd.DataFrame(ss.fit_transform(Xcon),columns=con)
```

```python
Xcon
```

# Joining Data

```python
X=Xcon.join(Xcat)
X
```

# Outliers Treatment

```python
out=[]
for i in Xcon.columns:
    ol=Xcon[(Xcon[i]>3)|(Xcon[i]<-3)].index
    out.extend(ol)
```

```python
outliers=list(set(out))
outliers
```

# Droping Outliers

```python
X=X.drop(index=outliers,axis=0)
Y=Y.drop(index=outliers,axis=0)
```

```python
# Reindexing
X.index=range(0,X.shape[0])
Y.index=range(0,Y.shape[0])
```

# Dividing Training and Testing Data

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=42)
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
model=lr.fit(xtrain,ytrain)
```

```python
tr_pred=model.predict(xtrain)
ts_pred=model.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score

tr_score=r2_score(tr_pred,ytrain)
ts_score=r2_score(ts_pred,ytest)
```

```python
print(tr_score)
print(ts_score)
```

```python
pred=model.predict(xtest)
error=mean_absolute_error(pred,ytest)
error
```

# Model Building Ridge

```python
from sklearn.linear_model import Ridge
rr=Ridge(alpha=10,random_state=1)
model=rr.fit(xtrain,ytrain)
tr_pred=model.predict(xtrain)
ts_pred=model.predict(xtest)


tr_score=r2_score(ytrain,tr_pred)
ts_score=r2_score(ytest,ts_pred)

print('training score',tr_score)
print('testing score',ts_score)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

param_grid = {'n_estimators': [2, 10, 20],'max_depth': [None, 5, 10],'min_samples_split': [2, 5],}
rf = RandomForestRegressor()

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='r2')
grid_search.fit(xtrain, ytrain)

best_rf = grid_search.best_estimator_


ts_pred = best_rf.predict(xtest)
tr_pred = best_rf.predict(xtrain)


tr_score = r2_score(tr_pred, ytrain)
ts_score = r2_score(ts_pred, ytest)


print(best_rf)
print('Training score:',tr_score)
print('Testing score:',ts_score)
```
